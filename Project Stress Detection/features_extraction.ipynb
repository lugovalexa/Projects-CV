## Reading additional files with TEMP and BVP

```python
temp = pd.DataFrame()
folders = glob.glob('./Empatica E4/*')
for i in folders:
    people_days = glob.glob(f'./{i}/*')
    for day in people_days:
        participantID = day.split("/")[3]
        temp_df = flirt.reader.empatica.read_temp_file_into_df(day +'/TEMP.csv')
        temp_df = temp_df.resample('1s', base=1).mean()
        temp_df['ID'] = participantID
        temp = temp.append(temp_df)
```

## Extracting BVP and TEMP features

```python
q = []
people = df['ID'].unique()
header = list(df.columns)

for p in people:
    q.append(
        [
            df[df['ID'] == p].head(1).index.values[0],
            df[df['ID'] == p].tail(1).index.values[0]
        ]
            )

del df
```

```python
def read_part_file(path,num):
    data = pd.read_csv(path, skiprows=q[num][0], nrows=q[num][1] - q[num][0] + 1)
    data = data.set_index(data.columns[0],
                        drop = True)
    data.index.name = None
    data.columns = header
    data = data.sort_values(by = 'datetime')

    return data
```

```python
temp_bvp = pd.DataFrame()
for num_part in range(len(q)):
    main_data = read_part_file('df_less_features.csv', num_part)
    main_data['date'] = pd.to_datetime(main_data['datetime']).dt.date
    main_data['date'] = pd.to_datetime(main_data['date'])

    for day in main_data.date.unique():
        data = main_data.loc[main_data.date == day].reset_index(drop=True)
        data = data[['datetime','temp','bvp','ID']]
        data['bvp_mean'] = 0.
        data['bvp_min'] = 0.
        data['bvp_max'] = 0.
        data['bvp_std'] = 0.
        data['bvp_peaks'] = 0.
        data['bvp_sign_changes'] = 0.
        data['temp_mean'] = 0.
        data['temp_min'] = 0.
        data['temp_max'] = 0.
        data['temp_std'] = 0.
        
        for i in range(len(data)):
            if i <= len(data) - 60:
                data['bvp_mean'][i] = data['bvp'][i:i+60].mean()
                data['bvp_min'][i] = data['bvp'][i:i+60].min()
                data['bvp_max'][i] = data['bvp'][i:i+60].max()
                data['bvp_std'][i] = data['bvp'][i:i+60].std()
                data['bvp_peaks'][i] = len(find_peaks(data['bvp'][i:i+60], prominence=0.9)[0])
                data['bvp_sign_changes'][i] = np.sum(np.diff(np.sign(data['bvp'][i:i+60])) != 0)

                data['temp_mean'][i] = data['temp'][i:i+60].mean()
                data['temp_min'][i] = data['temp'][i:i+60].min()
                data['temp_max'][i] = data['temp'][i:i+60].max()
                data['temp_std'][i] = data['temp'][i:i+60].std()
            else:
                data['bvp_mean'][i] = data['bvp'][len(data)-60:len(data)].mean()
                data['bvp_min'][i] = data['bvp'][len(data)-60:len(data)].min()
                data['bvp_max'][i] = data['bvp'][len(data)-60:len(data)].max()
                data['bvp_std'][i] = data['bvp'][len(data)-60:len(data)].std()
                data['bvp_peaks'][i] = len(find_peaks(data['bvp'][len(data)-60:len(data)], prominence=0.9)[0])
                data['bvp_sign_changes'][i] = np.sum(np.diff(np.sign(data['bvp'][len(data)-60:len(data)])) != 0)

                data['temp_mean'][i] = data['temp'][len(data)-60:len(data)].mean()
                data['temp_min'][i] = data['temp'][len(data)-60:len(data)].min()
                data['temp_max'][i] = data['temp'][len(data)-60:len(data)].max()
                data['temp_std'][i] = data['temp'][len(data)-60:len(data)].std()
        data = data.drop(columns=['temp','bvp'])
        temp_bvp = temp_bvp.append(data,ignore_index=True)
    del main_data
    temp_bvp.to_csv(f'temp_bvp_{num_part}.csv', index=False)
temp_bvp.to_csv('temp_bvp.csv', index=False)
```

## Adding tags and creating pre-stress and post-stress periods

```python
df
```

```python
tags = pd.read_csv('tags.csv', delimiter = ';', parse_dates = ['datetime'])
```

```python
df = df.drop(columns = 'tag')
df = pd.merge(df, tags, on=['ID','datetime'], how='left')
df = df.fillna(0)
df = df.drop_duplicates().reset_index(drop=True)
df['tag'] = df['tag'].astype('int64')
```

```python
df['date'] = pd.to_datetime(df['datetime']).dt.date
```

```python
tags_new = pd.DataFrame()
for id in df.ID.unique():
    data = df.loc[df.ID == id]
    for day in data.date.unique():
        data_day = data.loc[data.date == day].reset_index()
        data_day = data_day[['datetime','ID','tag']]
        i = 0
        while i <= len(data_day)-1:
            if data_day['tag'][i] == 1:
                data_day['tag'][i-1199:i] = 1
                data_day['tag'][i:i+1201] = 2
                i = i + 2401
            else:
                i += 1
        tags_new = tags_new.append(data_day, ignore_index=True)
```

```python
tags_new['date'] = pd.to_datetime(tags_new['datetime']).dt.date
tags_with_3 = pd.DataFrame()
for id in tags_new.ID.unique():
    data = tags_new.loc[tags_new.ID == id]
    if len(data.loc[data.tag == 1]) == 0:
        tags_with_3 = tags_with_3.append(data, ignore_index=True)
    else:
        for day in data.date.unique():
            data_day = data.loc[data.date == day].reset_index()
            data_day = data_day[['datetime','ID','tag']]
            length = len(data_day)
            a = 0
            while a <= length - 1:
                if sum(data_day['tag'][a:a+1]) == 0:
                    b = a
                    while sum(data_day['tag'][b:b+1]) == 0 and b <= length - 1:
                        b += 1
                    if b - a >= 7200:
                        data_day['tag'][a+3600:b-3599] = 3
                        a = b
                    else:
                        a = b
                else:
                    a += 1
            tags_with_3 = tags_with_3.append(data_day, ignore_index=True)
tags_with_3.tag.value_counts()
```

```python
tags_with_3 = tags_with_3.drop(columns = 'date')
```

```python
df = df.drop(columns = 'tag')
df = pd.merge(df, tags_with_3, on=['ID','datetime'],how='left')
```
